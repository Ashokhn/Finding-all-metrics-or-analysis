# Graphing the number of cofirmed cases,active cases,death cases,recovered cases


```python
#importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset

df=pd.read_csv('C:\\Users\\aravind\\Desktop\\VUMONIC DATALABS\\csse_daily.csv')
```

```python
df
```

```python
df.head()
```

```python
graphing=df.iloc[:,7:11]  #printing only the required fields/features for graphing
print(graphing)
```

```python
df.describe()  #describimg the given dataset
```

```python
df.isnull().sum()  #checking for null values from the dataset
```

```python
#graphing the corelation between the different features using heatmap

plt.figure(figsize=[10,5])
sns.heatmap(graphing.corr(),annot=True)
plt.show()
```

```python
#graphing various features

sns.pairplot(graphing)
```

```python
#finding the sum for the data 

df.sum(axis = 0, skipna = True)
```

```python
#finding the sum for the required features

confirmed_sum = df['Confirmed'].sum()
print(confirmed_sum)
deaths_sum = df['Deaths'].sum()
print(deaths_sum)
Recovered_sum = df['Recovered'].sum()
print(Recovered_sum)
Active_sum = df['Active'].sum()
print(Active_sum)
```

```python
# Data to plot
# use the method plt.pie() to create a plot.

labels = 'confirmed_sum', 'deaths_sum', 'Recovered_sum', 'Active_sum'
sizes = [12910357, 569128, 7116957, 5269635.0]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes,explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
#plot usinf bar chart


data = {'confirmed_sum':12910357, 'deaths_sum':569128, 'Recovered_sum':7116957,  
        'Active_sum':5269635.0} 
features = list(data.keys()) 
values = list(data.values()) 
   
fig = plt.figure(figsize = (10, 4)) 
  
# creating the bar plot 
plt.bar(features, values, color ='maroon',  
        width = 0.4) 
  
plt.xlabel("features") 
plt.ylabel("numbers") 
plt.title("pandermic details") 
plt.show()
```
